# Definición del problema

## Función de Rosenbrock

```python
def rosenbrock(x):
    return np.sum(100.0 * (x[1:] - x[:-1]**2.0)**2.0 + (1 - x[:-1])**2.0)
```

## Función de Rastrigin

```python
def rastrigin(x):
    A = 10
    n = len(x)
    return A * n + np.sum(x**2 - A * np.cos(2 * np.pi * x))
```

# 2. Optimización Tradicional: Descenso por Gradiente

## Prompt utilizado

Necesito implementar el algoritmo de descenso por gradiente en Python para optimizar las funciones de Rosenbrock y Rastrigin en 2 y 3 dimensiones. ¿Me ayudas a escribir una función base que reciba la función objetivo, su gradiente y devuelva tanto el punto óptimo como el historial de las posiciones visitadas (para luego hacer una animación)?

## Algoritmo Base de Descenso por Gradiente

```python
def descenso_gradiente(func, grad_func, dim, limites, lr=0.001, max_iter=1000, tol=1e-6):
    """
    Algoritmo base de Descenso por Gradiente
    """
    # Condición inicial aleatoria dentro de los límites
    x = np.random.uniform(limites[0], limites[1], dim)

    # Guardamos el historial para las futuras animaciones (Punto 4)
    historia = [x.copy()]

    for i in range(max_iter):
        gradiente = grad_func(x)

        # Actualización de la posición
        x_nuevo = x - lr * gradiente
        historia.append(x_nuevo.copy())

        # Criterio de parada (si el cambio es minúsculo)
        if np.linalg.norm(x_nuevo - x) < tol:
            break

        x = x_nuevo

    return x, func(x), np.array(historia), i + 1
```

## Rosenbrock

```python
# 1. Gradiente de Rosenbrock
def grad_rosenbrock(x):
    grad = np.zeros_like(x)
    grad[:-1] += -400 * x[:-1] * (x[1:] - x[:-1]**2) - 2 * (1 - x[:-1])
    grad[1:] += 200 * (x[1:] - x[:-1]**2)
    return grad

# Límites recomendados para Rosenbrock
limites_ros = (-2.048, 2.048)

# 2. Experimentos
print("--- DESCENSO POR GRADIENTE: ROSENBROCK 2D ---")
x_opt, f_opt, hist_ros2d, iteraciones = descenso_gradiente(rosenbrock, grad_rosenbrock, dim=2, limites=limites_ros, lr=0.001)
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt:.4f}")
print(f"Evaluaciones (iteraciones): {iteraciones}\n")

print("--- DESCENSO POR GRADIENTE: ROSENBROCK 3D ---")
x_opt, f_opt, hist_ros3d, iteraciones = descenso_gradiente(rosenbrock, grad_rosenbrock, dim=3, limites=limites_ros, lr=0.001)
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt:.4f}")
print(f"Evaluaciones (iteraciones): {iteraciones}")
```

## Rastrigin

```python
# 1. Gradiente de Rastrigin
def grad_rastrigin(x):
    return 2 * x + 20 * np.pi * np.sin(2 * np.pi * x)

# Límites recomendados para Rastrigin
limites_ras = (-5.12, 5.12)

# 2. Experimentos
print("--- DESCENSO POR GRADIENTE: RASTRIGIN 2D ---")
x_opt, f_opt, hist_ras2d, iteraciones = descenso_gradiente(rastrigin, grad_rastrigin, dim=2, limites=limites_ras, lr=0.001)
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt:.4f}")
print(f"Evaluaciones (iteraciones): {iteraciones}\n")

print("--- DESCENSO POR GRADIENTE: RASTRIGIN 3D ---")
x_opt, f_opt, hist_ras3d, iteraciones = descenso_gradiente(rastrigin, grad_rastrigin, dim=3, limites=limites_ras, lr=0.001)
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt:.4f}")
print(f"Evaluaciones (iteraciones): {iteraciones}")
```

## Optimización por Enjambre de Partículas (PSO)

```python
import numpy as np

# Definimos una función de wrapper para adaptar la salida y contar evaluaciones
evaluaciones_pso = 0
def objective_pso_wrapper(x, func):
    global evaluaciones_pso
    evaluaciones_pso += 1
    return func(x)

# Función auxiliar para extraer solo la mejor trayectoria (Forma: 150, 2)
def obtener_mejor_trayectoria(pso_model):
    mejor_hist_x = []
    mejor_y = float('inf')
    mejor_x = None
    for X_iter, Y_iter in zip(pso_model.record_value['X'], pso_model.record_value['Y']):
        Y_iter_flat = np.array(Y_iter).flatten()
        min_idx = np.argmin(Y_iter_flat)
        if Y_iter_flat[min_idx] < mejor_y:
            mejor_y = Y_iter_flat[min_idx]
            mejor_x = X_iter[min_idx].copy()
        mejor_hist_x.append(mejor_x)
    return np.array(mejor_hist_x)

# 1. Optimización de Rosenbrock
func_ros = lambda x: objective_pso_wrapper(x, rosenbrock)

print("--- ENJAMBRE DE PARTÍCULAS (PSO): ROSENBROCK 2D ---")
evaluaciones_pso = 0
# n_particles es el tamaño del enjambre
pso_2d = PSO(func=func_ros, n_dim=2, pop=40, max_iter=150, lb=limites_ros[0], ub=limites_ros[1])
pso_2d.record_mode = True
x_opt, f_opt = pso_2d.run()

# Extraemos el historial de la mejor partícula para la gráfica 3D
hist_pso_ros = obtener_mejor_trayectoria(pso_2d)

print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt[0]:.6f}")
print(f"Evaluaciones totales: {evaluaciones_pso}\n")

print("--- ENJAMBRE DE PARTÍCULAS (PSO): ROSENBROCK 3D ---")
evaluaciones_pso = 0
pso_3d = PSO(func=func_ros, n_dim=3, pop=40, max_iter=150, lb=limites_ros[0], ub=limites_ros[1])
x_opt, f_opt = pso_3d.run()
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt[0]:.6f}")
print(f"Evaluaciones totales: {evaluaciones_pso}\n")

print("-" * 20)

# 2. Optimización de Rastrigin
func_ras = lambda x: objective_pso_wrapper(x, rastrigin)

print("--- ENJAMBRE DE PARTÍCULAS (PSO): RASTRIGIN 2D ---")
evaluaciones_pso = 0
pso_2d = PSO(func=func_ras, n_dim=2, pop=40, max_iter=150, lb=limites_ras[0], ub=limites_ras[1])
pso_2d.record_mode = True
x_opt, f_opt = pso_2d.run()

# Extraemos el historial de la mejor partícula para la gráfica 3D
hist_pso_ras = obtener_mejor_trayectoria(pso_2d)

# ¡Este también debería ser casi 0!
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt[0]:.6f}")
print(f"Evaluaciones totales: {evaluaciones_pso}\n")

print("--- ENJAMBRE DE PARTÍCULAS (PSO): RASTRIGIN 3D ---")
evaluaciones_pso = 0
pso_3d = PSO(func=func_ras, n_dim=3, pop=40, max_iter=150, lb=limites_ras[0], ub=limites_ras[1])
x_opt, f_opt = pso_3d.run()
print(f"Punto final: {x_opt}")
print(f"Valor de la función: {f_opt[0]:.6f}")
print(f"Evaluaciones totales: {evaluaciones_pso}")
```

## Evolución Diferencial

```python
# Definimos una función de wrapper para adaptar la salida y contar evaluaciones
evaluaciones_de = 0
def objective_de_wrapper(x, func):
    global evaluaciones_de
    evaluaciones_de += 1
    return func(x)

# 1. Optimización de Rosenbrock
func_ros = lambda x: objective_de_wrapper(x, rosenbrock)
# Bounds formato para SciPy: [(min, max), (min, max), ...]
bounds_ros_2d = [limites_ros] * 2
bounds_ros_3d = [limites_ros] * 3

print("--- EVOLUCIÓN DIFERENCIAL: ROSENBROCK 2D ---")
evaluaciones_de = 0
# Ejecutamos differential_evolution
result = differential_evolution(func_ros, bounds_ros_2d, strategy='best1bin', maxiter=1000, popsize=15)
print(f"Punto final: {result.x}")
print(f"Valor de la función: {result.fun:.6f}")
print(f"Evaluaciones totales: {evaluaciones_de}\n")

print("--- EVOLUCIÓN DIFERENCIAL: ROSENBROCK 3D ---")
evaluaciones_de = 0
result = differential_evolution(func_ros, bounds_ros_3d, strategy='best1bin', maxiter=1000, popsize=15)
print(f"Punto final: {result.x}")
print(f"Valor de la función: {result.fun:.6f}")
print(f"Evaluaciones totales: {evaluaciones_de}\n")

print("-" * 20)

# 2. Optimización de Rastrigin
func_ras = lambda x: objective_de_wrapper(x, rastrigin)
bounds_ras_2d = [limites_ras] * 2
bounds_ras_3d = [limites_ras] * 3

print("--- EVOLUCIÓN DIFERENCIAL: RASTRIGIN 2D ---")
evaluaciones_de = 0
result = differential_evolution(func_ras, bounds_ras_2d, strategy='best1bin', maxiter=1000, popsize=15)
# ¡Este es súper robusto, debe dar 0 exacto!
print(f"Punto final: {result.x}")
print(f"Valor de la función: {result.fun:.6f}")
print(f"Evaluaciones totales: {evaluaciones_de}\n")

print("--- EVOLUCIÓN DIFERENCIAL: RASTRIGIN 3D ---")
evaluaciones_de = 0
result = differential_evolution(func_ras, bounds_ras_3d, strategy='best1bin', maxiter=1000, popsize=15)
print(f"Punto final: {result.x}")
print(f"Valor de la función: {result.fun:.6f}")
print(f"Evaluaciones totales: {evaluaciones_de}")
```

# 4. Visualización de Trayectorias

## Motores de Visualización (Código Base)

### Prompt utilizado

Genera funciones en Python utilizando Matplotlib para visualizar procesos de optimización en 2D y 3D. La función 2D debe crear una animación de contorno (contour plot) que muestre el movimiento paso a paso. La función 3D debe generar una superficie estática (surface plot) de alta calidad que permita ver la topografía de la función y la trayectoria completa del algoritmo sobre ella.

### Código Base para animaciones

```python
def crear_animacion_contorno(func, historial, limites, titulo):
    fig, ax = plt.subplots(figsize=(8, 6))
    x = np.linspace(limites[0], limites[1], 100)
    y = np.linspace(limites[0], limites[1], 100)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[func(np.array([xi, yi])) for xi in x] for yi in y])

    contorno = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
    plt.colorbar(contorno)
    linea, = ax.plot([], [], 'r-', lw=2, label='Trayectoria')
    punto, = ax.plot([], [], 'ro')

    def init():
        ax.set_title(titulo)
        linea.set_data([], [])
        punto.set_data([], [])
        return linea, punto

    def update(frame):
        data = historial[:frame]
        linea.set_data(data[:, 0], data[:, 1])
        punto.set_data([historial[frame-1, 0]], [historial[frame-1, 1]])
        return linea, punto

    anim = FuncAnimation(fig, update, frames=len(historial), init_func=init, blit=True, interval=50)
    plt.close()
    return anim

def graficar_trayectoria_3d(func, historial, limites, titulo):
    x = np.linspace(limites[0], limites[1], 50)
    y = np.linspace(limites[0], limites[1], 50)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[func(np.array([xi, yi])) for xi in x] for yi in y])

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)

    z_hist = np.array([func(p) for p in historial])
    ax.plot(historial[:, 0], historial[:, 1], z_hist, 'r-', lw=3, zorder=10)
    ax.scatter(historial[-1, 0], historial[-1, 1], z_hist[-1], color='black', s=50)

    ax.set_title(titulo)
    plt.show()
```

# Discusión de Resultados

## Análisis Estadístico (Múltiples Corridas)

Dado el componente estocástico de los algoritmos implementados (la inicialización aleatoria en el Descenso por Gradiente y la naturaleza probabilística de los operadores en los métodos heurísticos), una sola ejecución carece de validez estadística.

Para obtener conclusiones rigurosas, se realizaron **30 corridas independientes** para cada algoritmo sobre las funciones de prueba en 2D. Esto permite extraer el promedio y la desviación estándar tanto del **valor final de la función objetivo** como del **número de evaluaciones**, garantizando que la discusión se fundamente en el comportamiento promedio y la robustez del método, y no en casos aislados de éxito o fracaso.

```python
import numpy as np

num_corridas = 30
print(f"=== ESTADÍSTICAS DE {num_corridas} CORRIDAS INDEPENDIENTES (2D) ===")

def imprimir_stats(nombre, resultados, evaluaciones):
    print(f"\n[{nombre}]")
    print(f"Mejor valor: {np.min(resultados):.6f}")
    print(f"Peor valor:  {np.max(resultados):.6f}")
    print(f"Promedio:    {np.mean(resultados):.6f} (+/- {np.std(resultados):.6f})")
    print(f"Evals prom:  {np.mean(evaluaciones):.1f}")

# ==========================================
# 1. DESCENSO POR GRADIENTE (DG)
# ==========================================
res_dg_ros, evals_dg_ros = [], []
res_dg_ras, evals_dg_ras = [], []

for _ in range(num_corridas):
    # Rosenbrock (Le bajamos el lr a 0.0001 para que no explote)
    _, f_opt_ros, _, iter_ros = descenso_gradiente(rosenbrock, grad_rosenbrock, 2, limites_ros, lr=0.0001)
    res_dg_ros.append(f_opt_ros)
    evals_dg_ros.append(iter_ros)

    # Rastrigin
    _, f_opt_ras, _, iter_ras = descenso_gradiente(rastrigin, grad_rastrigin, 2, limites_ras, lr=0.001)
    res_dg_ras.append(f_opt_ras)
    evals_dg_ras.append(iter_ras)

imprimir_stats("DG - Rosenbrock", res_dg_ros, evals_dg_ros)
imprimir_stats("DG - Rastrigin", res_dg_ras, evals_dg_ras)


# ==========================================
# 2. ALGORITMO GENÉTICO (GA)
# ==========================================
res_ga_ros, evals_ga_ros = [], []
res_ga_ras, evals_ga_ras = [], []

for _ in range(num_corridas):
    # Rosenbrock
    evaluaciones_ga = 0
    func_ros_ga = lambda x: objective_ga_wrapper(x, rosenbrock)
    ga_ros = GA(func=func_ros_ga, n_dim=2, size_pop=50, max_iter=200, lb=limites_ros[0], ub=limites_ros[1])
    _, f_opt_ros = ga_ros.run()
    res_ga_ros.append(f_opt_ros[0])
    evals_ga_ros.append(evaluaciones_ga)

    # Rastrigin
    evaluaciones_ga = 0
    func_ras_ga = lambda x: objective_ga_wrapper(x, rastrigin)
    ga_ras = GA(func=func_ras_ga, n_dim=2, size_pop=50, max_iter=200, lb=limites_ras[0], ub=limites_ras[1])
    _, f_opt_ras = ga_ras.run()
    res_ga_ras.append(f_opt_ras[0])
    evals_ga_ras.append(evaluaciones_ga)

imprimir_stats("GA - Rosenbrock", res_ga_ros, evals_ga_ros)
imprimir_stats("GA - Rastrigin", res_ga_ras, evals_ga_ras)


# ==========================================
# 3. ENJAMBRE DE PARTÍCULAS (PSO)
# ==========================================
res_pso_ros, evals_pso_ros = [], []
res_pso_ras, evals_pso_ras = [], []

for _ in range(num_corridas):
    # Rosenbrock
    evaluaciones_pso = 0
    func_ros_pso = lambda x: objective_pso_wrapper(x, rosenbrock)
    pso_ros = PSO(func=func_ros_pso, n_dim=2, pop=40, max_iter=150, lb=limites_ros[0], ub=limites_ros[1])
    _, f_opt_ros = pso_ros.run()
    res_pso_ros.append(f_opt_ros[0])
    evals_pso_ros.append(evaluaciones_pso)

    # Rastrigin
    evaluaciones_pso = 0
    func_ras_pso = lambda x: objective_pso_wrapper(x, rastrigin)
    pso_ras = PSO(func=func_ras_pso, n_dim=2, pop=40, max_iter=150, lb=limites_ras[0], ub=limites_ras[1])
    _, f_opt_ras = pso_ras.run()
    res_pso_ras.append(f_opt_ras[0])
    evals_pso_ras.append(evaluaciones_pso)

imprimir_stats("PSO - Rosenbrock", res_pso_ros, evals_pso_ros)
imprimir_stats("PSO - Rastrigin", res_pso_ras, evals_pso_ras)


# ==========================================
# 4. EVOLUCIÓN DIFERENCIAL (DE)
# ==========================================
res_de_ros, evals_de_ros = [], []
res_de_ras, evals_de_ras = [], []
bounds_ros = [limites_ros] * 2
bounds_ras = [limites_ras] * 2

for _ in range(num_corridas):
    # Rosenbrock
    evaluaciones_de = 0
    func_ros_de = lambda x: objective_de_wrapper(x, rosenbrock)
    res_ros = differential_evolution(func_ros_de, bounds_ros, strategy='best1bin', maxiter=1000, popsize=15)
    res_de_ros.append(res_ros.fun)
    evals_de_ros.append(evaluaciones_de)

    # Rastrigin
    evaluaciones_de = 0
    func_ras_de = lambda x: objective_de_wrapper(x, rastrigin)
    res_ras = differential_evolution(func_ras_de, bounds_ras, strategy='best1bin', maxiter=1000, popsize=15)
    res_de_ras.append(res_ras.fun)
    evals_de_ras.append(evaluaciones_de)

imprimir_stats("DE - Rosenbrock", res_de_ros, evals_de_ros)
imprimir_stats("DE - Rastrigin", res_de_ras, evals_de_ras)
```

## Aportes del Descenso por Gradiente (DG)

El Descenso por Gradiente demostró ser altamente dependiente de la topología de la función y de su punto de inicio aleatorio, mostrando una extrema fragilidad en escenarios complejos:
* **Convergencia prematura (Rastrigin):** Al ser un método puramente local, el DG quedó atrapado sistemáticamente en los mínimos locales generados por los términos cosenoidales de Rastrigin. Esto se refleja en un promedio de evaluaciones sumamente bajo (27.9 iteraciones), deteniéndose no por encontrar el óptimo, sino por anulación del gradiente. Entregó un valor promedio muy deficiente (18.97) con una desviación estándar enorme (+/- 10.08), demostrando que su éxito es cuestión de azar.
* **Lentitud en valles planos (Rosenbrock):** En esta función, el DG agotó el límite máximo de iteraciones permitidas (1000 iteraciones promedio) rebotando en el valle estrecho. Aunque se acercó al óptimo (promedio 0.99), fue incapaz de hacer la sintonía fina final de manera eficiente.

## Aportes de los Métodos Heurísticos (AG, PSO, DE)

Los algoritmos poblacionales justificaron su uso al lograr lo que el DG no pudo: evadir trampas locales y encontrar el mínimo global, aunque con comportamientos distintos entre ellos:
* **Algoritmo Genético (GA):** Demostró una gran capacidad para acercarse a la región óptima, pero exhibió dificultades para la "sintonía fina" en un espacio continuo. Agotó siempre su presupuesto máximo de evaluaciones (10001.0) y, aunque logró promedios bajos (0.07 en Rosenbrock y 0.69 en Rastrigin), no alcanzó el cero perfecto, mostrando la limitación de sus operadores clásicos de mutación y cruce para precisiones decimales extremas.
* **Optimización por Enjambre de Partículas (PSO):** Mostró una convergencia perfecta y una robustez absoluta. En las 30 corridas logró un valor de **0.000000 (+/- 0.0)** en ambas funciones. Su estrategia fue de "fuerza bruta eficiente", evaluando siempre a toda la población durante todas las generaciones estipuladas (6040 evaluaciones estáticas).
* **Evolución Diferencial (DE) - El balance óptimo:** DE resultó ser el algoritmo superior en este experimento. Al igual que el PSO, logró el óptimo global perfecto (**0.000000**) sin ninguna desviación estándar, pero lo hizo con un mecanismo de parada dinámico mucho más inteligente. Requirió solo **3990 evaluaciones en Rosenbrock** y apenas **1987 evaluaciones en Rastrigin** para garantizar la convergencia, superando ampliamente la eficiencia computacional del PSO y del GA.

## Balance: Valor Final vs. Evaluaciones de la Función Objetivo

¿Qué aportó cada método? El **Descenso por Gradiente** aporta una explotación local veloz, útil únicamente si el espacio es convexo o si la región del mínimo es conocida a priori. Por su parte, los **métodos heurísticos (destacando la Evolución Diferencial)** aportan la exploración global necesaria para resolver topologías multimodales (Rastrigin) y valles degenerados (Rosenbrock).

Los resultados confirman que el incremento en el número de evaluaciones de la función objetivo (pasar de decenas a miles) no es un desperdicio de recursos, sino el costo computacional necesario y justificado para garantizar la convergencia al óptimo global en problemas matemáticos complejos.
